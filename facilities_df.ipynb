```python
len(df3.index)
```

```python
weight_rank = []

for i in range(0,len(df3)):
    item1 = df3.iloc[i,0]
    item2 = df4.iloc[i,2]
    weight_rank.append(item1*item2/1000)

data = {'State': df3.index, 'Weighted Rank': weight_rank}    

```
